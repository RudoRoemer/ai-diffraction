```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
dataset = os.path.join("c:\\users\\joe\\desktop\\ai-diffraction\\datasets\\lacbed")
```

```python
n = 20
```

```python
output = np.zeros((2 * 128, n * 128))
```

```python
codes = np.random.choice(os.listdir(dataset), n)

for i, code in enumerate(codes):
    im = cv2.imread(
        os.path.join(dataset, code, code + "_structure.png")
    )
    im = im[:,:,0]
    output[0: 128, i * 128: (i + 1) * 128] = im
    im2 = cv2.imread(
        os.path.join(dataset, code, code + "_+0+0+0.png")
    )
    im2 = im2[:,:,0]
    output[128: 256, i * 128: (i + 1) * 128] = im2

plt.figure(figsize=(20, 2))
plt.imshow(output)
print(codes)
```
