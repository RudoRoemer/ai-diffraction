```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
dataset = os.path.join("..\\datasets\\lacbed")
results = os.path.join("..\\results\\pattern_random0\\val_100\\images")
results2 = os.path.join("..\\results\\pattern_random0\\test_100\\images")
```

```python
codes = np.random.choice(os.listdir(results), 10)
codes = [code.split("_")[0] for code in codes]
print(codes)
# codes = [237494, 83326, 152259, 391131, 156497, 608458, 86352, 169152, 54390]
# codes = ['83004', '59179', '65341', '84138', '415604', '405958', '43724', '426373', '291613', '49554', '426098', '20259', '45818', '196285', '100860', '105431', '165394', '8749', '94023', '169863']
# codes = [391131, 152259, 54390]
# codes = [152794, 162198, 166923, 194751]

codes = [str(code) for code in codes]

output = np.zeros((3 * 128, len(codes) * 128))

for i, code in enumerate(codes):
    print()
    im = cv2.imread(
        os.path.join(dataset, code, code + "+0+0+0.png")
    )
    im = im[:,:,0]
    output[0: 128, i * 128: (i + 1) * 128] = im
    im2 = cv2.imread(
        os.path.join(dataset, code, code + "_structure.png")
    )
    im2 = im2[:,:,0]
    output[128: 256, i * 128: (i + 1) * 128] = im2
    try:
        im3 = cv2.imread(
            os.path.join(results, code + "_structure_synthesized_image.png")
        )
        # im3 = im3[:,:,0]
        output[256: 384, i * 128: (i + 1) * 128] = im3
    except:
        im3 = cv2.imread(
            os.path.join(results2, code + "_structure_synthesized_image.png")
        )
        im3 = im3[:,:,0]
        output[256: 384, i * 128: (i + 1) * 128] = im3

plt.figure(figsize=(len(codes), 3))
plt.imshow(output)
print(codes)
```
